* El sistema deberá ofrecer cuatro opciones: calcular el área de un círculo, calcular su perímetro, calcular el volumen de una esfera (utilizando el mismo concepto de radio) o salir del programa. Cada operación solicitará el valor del radio únicamente cuando se ejecute, aplicará la fórmula matemática correspondiente y mostrará el resultado con dos decimales. El menú se mantendrá activo mediante un ciclo repetitivo hasta que el usuario decida finalizar, utilizando funciones modulares y estructuras de control (while, if, elif, else) para organizar el código de forma clara, sin emplear programación orientada a objetos ni validación de datos.

1. A = π × r²
2. P = 2 × π × r
3. V = (4/3) × π × r³

```python
# ======================
# PROGRAMACION DE LOS CALCULOS GEOMETRICOS
# elaborado por ian gutierrez
# formulas area del circulo, perimetro del circulo, volumen de la esfera


import math


def areacirculo():
    """Calcula el área de un círculo"""
    print("\n📐 ── CALCULAR ÁREA DEL CÍRCULO ──")
    radio = float(input("🔹 Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"✅ Resultado: El área es {area:.2f} unidades²\n")

def perimetrocirculo():
    """Calcula el perímetro de un círculo"""
    print("\n📏 ── CALCULAR PERÍMETRO DEL CÍRCULO ──")
    radio = float(input("🔹 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ Resultado: El perímetro es {perimetro:.2f} unidades\n")

def volumenesfera():
    """Calcula el volumen de una esfera"""
    print("\n🌍 ── CALCULAR VOLUMEN DE LA ESFERA ──")
    radio = float(input("🔹 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"✅ Resultado: El volumen es {volumen:.2f} unidades³\n")

def menu():
    """muestra el menu interactivo y controla el flujo del programa"""
    while True:
        print("\n" + "✨" * 20)
        print("🎓 MENÚ DE CÁLCULOS GEOMÉTRICOS 🎓")
        print("✨" * 20)
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")

        opcion = int(input("👉 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🌟")
            break
        else:
            print("⚠️ Opción no válida. Por favor, selecciona un numero del 1 al 4.")
            print("programa realizado por ian gutierrez")

# Ejecución principal
if __name__ == "__main__":
    menu()
```
